```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt
import seaborn as sns
import math

import folium
from folium.plugins import HeatMap

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#import and read dataframe
url = 'https://raw.githubusercontent.com/johannagan/california-wildfires/main/California_Fire_Incidents.csv'
wf = pd.read_csv(url)
wf.head()
```

```python
map = folium.Map(location=[36,-120], zoom_start=6)
for i in range(len(wf)):
    folium.Circle(
        location=[wf.loc[i,'Latitude'], wf.loc[i,'Longitude']],
        radius=math.sqrt(float(wf.loc[i,'AcresBurned'])*4047/3.14),
        popup=str(wf.loc[i,'Name'])+', '+str(wf.loc[i,'ArchiveYear']),
        color='red',
        fill=True,
        fill_color='red',
    ).add_to(map)
    
map
```
